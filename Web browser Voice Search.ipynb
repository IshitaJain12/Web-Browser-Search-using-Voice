```python
from tkinter import*
from tkinter import ttk
import webbrowser
import speech_recognition as sr
import pyaudio as p



root=Tk()
root.title("speech+vocab")
root.iconbitmap('C:\\Users\\HP\\Desktop\\mic.ico')
style=ttk.Style()
style.theme_use('winnative')
photo=PhotoImage(file='microphone.png').subsample(25,25)

labell1=Label(root,text="Query:")
labell1.grid(row=0,column=0)
entry1=Entry(root,width=40)
entry1.grid(row=0,column=1,columnspan=4)
btn2=StringVar()
def call_back():
    if(btn2.get()=="Google" and btn2.get()!=''):
        webbrowser.open('http://google.com/search?q='+entry1.get())
    elif(btn2.get()=="Duck" and btn2.get()!=''):
        webbrowser.open('http://duckduckgo.com/?q='+entry1.get())
    elif(btn2.get()=="Amz" and btn2.get()!=''):
        webbrowser.open('https://www.amazon.in/s/ref=nb_sb_noss_2?url=search-alias%3Daps&field-keywords='+entry1.get())
    elif(btn2.get()=="Ytb" and btn2.get()!=''):
         webbrowser.open('https://www.youtube.com/results?search_query='+entry1.get())
    else:
        pass
        
def get(event):
    if(btn2.get()=="Google" and btn2.get()!=''):
        webbrowser.open('http://google.com/search?q='+entry1.get())
    elif(btn2.get()=="Duck" and btn2.get()!=''):
        webbrowser.open('http://duckduckgo.com/?q='+entry1.get())
    elif(btn2.get()=="Amz" and btn2.get()!=''):
        webbrowser.open('https://www.amazon.in/s/ref=nb_sb_noss_2?url=search-alias%3Daps&field-keywords='+entry1.get())
    elif(btn2.get()=="Ytb" and btn2.get()!=''):
         webbrowser.open('https://www.youtube.com/results?search_query='+entry1.get())
    else:
        pass
def buttonClick():
    r=sr.Recognizer();
    r.energy_threshold=400
    r.pause_threshold=0.7
    with sr.Microphone() as source:
        print("Say Something");
        audio=r.listen(source,timeout=5);
        


    try:
        text=r.recognize_google(audio);
        print("You Said: ",format(text));
        entry1.focus()
        entry1.delete(0,END)
        entry1.insert(0,text)
        
        if(btn2.get()=="Google"):
            webbrowser.open('http://google.com/search?q='+text)
        elif(btn2.get()=="Duck"):
            webbrowser.open('http://duckduckgo.com/?q='+text)
        elif(btn2.get()=="Amz"):
            webbrowser.open('https://www.amazon.in/s/ref=nb_sb_noss_2?url=search-alias%3Daps&field-keywords='+text)
        elif(btn2.get()=="Ytb"):
            webbrowser.open('https://www.youtube.com/results?search_query='+entry1.get())
        else:
            pass
    except:
        print("Sorry could not recognise you")
    else:
        pass
entry1.bind('<Return>',get)

Mybutton1=Button(root,text="Search",width=10,command=call_back)
Mybutton1.grid(row=0,column=6)



Mybutton2=Radiobutton(root,text="Google",value='Google',variable=btn2)
Mybutton2.grid(row=1,column=1,sticky='w')

Mybutton3=Radiobutton(root,text="Duck",value='Duck',variable=btn2)
Mybutton3.grid(row=1,column=2,sticky='w')

Mybutton4=Radiobutton(root,text="Amz",value='Amz',variable=btn2)
Mybutton4.grid(row=1,column=3)

Mybutton5=Radiobutton(root,text="Ytb",value='Ytb',variable=btn2)
Mybutton5.grid(row=1,column=4,sticky='w')

Mybutton6=Button(root,image=photo,command=buttonClick,bd=0,overrelief='groove',relief='sunken',activebackground='gray')
Mybutton6.grid(row=0,column=5)
entry1.focus()
root.wm_attributes('-topmost',1)
root.mainloop()
    
    
```
